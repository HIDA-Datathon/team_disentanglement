# Model Differences

```python
%load_ext autoreload
%autoreload 2

import xarray as xr
import numpy as np
from plot_map import plot_map
from settings import path_data
```

```python
ds_solar = xr.open_dataset(path_data / 'Solar_forcing_1st_mill.nc')
ds_volc = xr.open_dataset(path_data / 'Volc_Forc_AOD_1st_mill.nc')
ds_temp_r1 = xr.open_dataset(path_data / 'T2m_R1_ym_1stMill.nc')
ds_temp_r2 = xr.open_dataset(path_data / 'T2m_R2_ym_1stMill.nc')
```

```python
mean_temp_r1 = np.mean(ds_temp_r1['T2m'].values, axis=0)
mean_temp_r1 = mean_temp_r1 / np.max(mean_temp_r1)
mean_temp_r2 = np.mean(ds_temp_r2['T2m'].values, axis=0)
mean_temp_r2 = mean_temp_r2 / np.max(mean_temp_r2)

plot_map(mean_temp_r1 - mean_temp_r2, title='Normalized Mean Difference', colorbar_title='r1 - r2')
```

```python
std_temp_r1 = np.std(ds_temp_r1['T2m'].values, axis=0)
std_temp_r1 = std_temp_r1 / np.max(std_temp_r1)
std_temp_r2 = np.std(ds_temp_r2['T2m'].values, axis=0)
std_temp_r2 = std_temp_r2 / np.max(std_temp_r2)

plot_map(std_temp_r1 - std_temp_r2, title='Normalized Std Difference', colorbar_title='r1 - r2')
```

Neither the mean nor the standard deviation of the temperature differs much between the models (the values on the colorbar are very low)
